```python
import cv2
import torch
import matplotlib.pyplot as plt
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO("yolov8n.pt")  # Use "yolov8s.pt" or "yolov8m.pt" for better accuracy

# Provide the image path
image_path = "pic.jpg"  # Replace with your image file path

# Perform inference
results = model(image_path)  # This returns a list of `Results` objects

# Get the first result
result = results[0]  

# ✅ Step 1: Show detections using `plot()`
annotated_image = result.plot()

# Convert image from BGR to RGB (for displaying correctly in matplotlib)
annotated_image_rgb = cv2.cvtColor(annotated_image, cv2.COLOR_BGR2RGB)

# Display the annotated image
plt.imshow(annotated_image_rgb)
plt.axis('off')
plt.show()

# ✅ Step 2: Extract bounding box coordinates, class names, and confidence scores
for box in result.boxes.data.tolist():
    x1, y1, x2, y2, confidence, class_id = box
    class_name = result.names[int(class_id)]
    print(f"Detected: {class_name} | Confidence: {confidence:.2f} | Bounding Box: ({x1}, {y1}, {x2}, {y2})")
```
